## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Make a copy of the original dataset
df_clean = df.copy()

# Drop unnecessary columns
df_clean.drop(columns=['ID', 'Name', 'SSN'], inplace=True)

# Drop duplicate rows
df_clean.drop_duplicates(inplace=True)

# Fill missing values
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].mean())

categorical_cols = df_clean.select_dtypes(include='object').columns
df_clean[categorical_cols] = df_clean[categorical_cols].fillna(df_clean[categorical_cols].mode().iloc[0])

# Fix data types if needed
df_clean['Month'] = df_clean['Month'].astype(int)
fix_cats = ['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Credit_Score']
df_clean[fix_cats] = df_clean[fix_cats].astype(str)

# Optional: Create a new feature
df_clean['Debt_Income_Ratio'] = df_clean['Outstanding_Debt'] / (df_clean['Annual_Income'] + 1)  # avoid div/0
````

### What all manipulations have you done and insights you found?

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothesis 1: Annual Income vs Credit Score

Hypothesis 2: Credit Utilization Ratio vs Credit Score

Hypothesis 3: Payment of Minimum Amount vs Credit Score

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀):
There is no significant difference in the mean Annual Income across different Credit Score categories (Good, Standard, Poor).

Alternate Hypothesis (H₁):
There is a significant difference in mean Annual Income across Credit Score categories.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value


from scipy.stats import f_oneway

# Group annual income by credit score
good = df_clean[df_clean['Credit_Score'] == 'Good']['Annual_Income']
standard = df_clean[df_clean['Credit_Score'] == 'Standard']['Annual_Income']
poor = df_clean[df_clean['Credit_Score'] == 'Poor']['Annual_Income']

# ANOVA Test
f_stat, p_val = f_oneway(good, standard, poor)
print("ANOVA P-Value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

We used the One-Way ANOVA test.

##### Why did you choose the specific statistical test?

Because we're comparing the means of a continuous variable (Annual_Income) across three independent categorical groups (Credit_Score) — Good, Standard, and Poor. ANOVA is the appropriate test for this case.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀):
There is no significant difference in the mean Credit Utilization Ratio between customers with Good and Poor credit scores.

Alternate Hypothesis (H₁):
There is a significant difference in mean Credit Utilization Ratio between Good and Poor credit score groups.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import ttest_ind

# Filter only Good and Poor score groups
good_util = df_clean[df_clean['Credit_Score'] == 'Good']['Credit_Utilization_Ratio']
poor_util = df_clean[df_clean['Credit_Score'] == 'Poor']['Credit_Utilization_Ratio']

# T-Test
t_stat, p_val = ttest_ind(good_util, poor_util, equal_var=False)
print("T-test P-Value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

We performed an Independent Samples t-test.

##### Why did you choose the specific statistical test?

Because we're comparing the means of a continuous variable (Credit_Utilization_Ratio) between two independent groups — customers with Good and Poor credit scores. A t-test is appropriate for two-group mean comparisons.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

There is no association between Payment of Minimum Amount (Yes/No) and Credit Score.

There is an association between Payment of Minimum Amount and Credit Score.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value


from scipy.stats import chi2_contingency
import pandas as pd

# Create contingency table
contingency = pd.crosstab(df_clean['Payment_of_Min_Amount'], df_clean['Credit_Score'])

# Chi-Square Test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency)
print("Chi-Square P-Value:", p_val)
````

## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

# Handling missing values for numerical columns
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].mean())

# Handling missing values for categorical columns
categorical_cols = df_clean.select_dtypes(include='object').columns
df_clean[categorical_cols] = df_clean[categorical_cols].fillna(df_clean[categorical_cols].mode().iloc[0])
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Mean imputation for numerical features to retain the overall average and reduce distortion in the data.

Mode imputation for categorical features to replace missing values with the most frequent category, which preserves the category distribution.
These are simple yet effective strategies for maintaining consistency and avoiding loss of data.



### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

# Cap outliers using IQR method (Winsorization style)
def cap_outliers(col):
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_clean[col] = df_clean[col].clip(lower, upper)

outlier_cols = ['Annual_Income', 'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_EMI_per_month']
for col in outlier_cols:
    cap_outliers(col)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

We applied the IQR (Interquartile Range) method to cap extreme outliers in numerical columns. This method helps:

Prevent skewing of the distribution by extreme values

Preserve data by not dropping rows

Make machine learning models more robust and less biased

### 3. Categorical Encoding

````python
# Encode your categorical columns

from sklearn.preprocessing import LabelEncoder

# Label encode ordinal-like categorical features
label_cols = ['Credit_Mix', 'Payment_of_Min_Amount', 'Credit_Score']
le = LabelEncoder()
for col in label_cols:
    df_clean[col] = le.fit_transform(df_clean[col])

# One-hot encode nominal categorical features
df_clean = pd.get_dummies(df_clean, columns=['Occupation', 'Payment_Behaviour'], drop_first=True)
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

Label Encoding was used for ordinal variables like Credit_Score, Credit_Mix, and Payment_of_Min_Amount, where the categories have a meaningful order.

One-Hot Encoding was used for nominal features like Occupation and Payment_Behaviour, where categories have no inherent order.

### 7. Dimesionality Reduction

##### Do you think that dimensionality reduction is needed? Explain Why?

Yes, dimensionality reduction is useful in this dataset because it contains a large number of features, including multiple one-hot encoded columns (like Occupation_*, Payment_Behaviour_*) that can lead to multicollinearity, overfitting, and high computational cost. Many features may also be correlated or redundant, which can reduce model generalization.

````python
# Drop all non-numeric columns (like 'Name', 'Type_of_Loan', etc.)
X_numeric = X.select_dtypes(include=['number'])  # Keep only numeric features

from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_numeric)

print(f"Original Features: {X_numeric.shape[1]}")
print(f"Reduced Features after PCA: {X_pca.shape[1]}")
````

##### Which dimensionality reduction technique have you used and why? (If dimensionality reduction done on dataset.)

I have used Principal Component Analysis (PCA).

Why PCA?
PCA is an unsupervised linear transformation technique that projects data into a lower-dimensional space while preserving maximum variance.

It is effective for handling correlated numerical features and helps improve model efficiency without significant loss in information.

In our case, PCA retained 95% of the variance while reducing the number of features, which indicates that most of the original information is preserved.
